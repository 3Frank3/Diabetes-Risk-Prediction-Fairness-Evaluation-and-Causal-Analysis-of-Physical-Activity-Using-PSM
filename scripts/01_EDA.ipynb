## Load Data

```python
from pathlib import Path
import pandas as pd
from IPython.display import display, Image
import diabetes_eda_extended_report as eda

eda.ensure_dirs()
eda.set_style()
df = eda.load_data()
print('Shape:', df.shape)
print('Columns:', list(df.columns))
```

## Create Analysis DataFrame

```python
section_df = df[eda.SECTION_VARS + ['BMI_group', 'PhysHlthPct30', 'MentHlthPct30']].copy()
display(section_df.head())
display(eda.summary_table(section_df[[c for c in section_df.columns if c in df.columns]]))
```
