# 03_Fairness

Standalone fairness runner for subgroup performance evaluation.

## Run fairness pipeline

```python
from IPython.display import display, Image
import importlib.util
from pathlib import Path

module_path = Path('/Users/zansong/Desktop/203C/Project/scripts/03_Fairness.py')
spec = importlib.util.spec_from_file_location('fairness_runner', module_path)
fairness_runner = importlib.util.module_from_spec(spec)
spec.loader.exec_module(fairness_runner)
```

```python
artifacts = fairness_runner.run_fairness_pipeline()
print('Model used:', artifacts['model_name'])
```

## Tables

```python
display(artifacts['prevalence_table'])
display(artifacts['metrics_table'])
display(artifacts['gap_table'][['Group', 'Subgroup', 'Recall_gap', 'FNR_gap', 'AUC_gap', 'Precision_gap']])
```

## Plots

```python
for p in artifacts['plot_paths']:
    display(Image(filename=str(p)))
```

## Output files

```python
print('Tables:')
for p in artifacts['table_paths']:
    print('-', p)
print('\nPlots:')
for p in artifacts['plot_paths']:
    print('-', p)
```
